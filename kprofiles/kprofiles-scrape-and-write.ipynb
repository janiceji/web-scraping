# Functions

```python
# CONVERTS LIST TO DICTIONARY (member_dic HELPER)
def convert(member_list):
    somelist = [x for x in member_list if ":" in x]
    return dict(element.split(':', 1) for element in somelist)
```

```python
# TRANSFORMS LIST OF LIST INTO LIST OF DICTIONARY
def member_dic(member_list):
    bosokham =[]
    for jewel in member_list:
        temp = convert(jewel)
        bosokham.append(temp)
    return bosokham
```

```python
# CLEAN THE PROFILE PAGE
def clean_members(p_rows):
    temp = []
    temp2 =[]
    for i in range(len(p_rows)):
        curr_row = p_rows[i]
        span = curr_row.findAll("span")
        for j in range(1, len(span)):
            try:
                if span[j].next_sibling != " ":
                    lst = temp.append(span[j].text + span[j].next_sibling)
                else:
                    pass       
            except:
                temp.append("na")
        temp = []
        temp2.append(temp)
    cleaned_members = [element for element in temp2 if element != []]
    return cleaned_members
```

```python
def dic_to_df(dic):
    df = pd.DataFrame.from_dict(dic)
    return df
```
